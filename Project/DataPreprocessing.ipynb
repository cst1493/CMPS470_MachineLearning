```python
import math
import numpy as np
import pandas as pd
import statistics as stats
from sklearn.model_selection import train_test_split

data = pd.read_csv('GoldWeekly2001-2021.csv')
data = pd.DataFrame(data)

class weekly_gold():
    def Date(self, date):
        self.Date = date
    
    def Open(self, Open):
        self.Open = Open
    
    def High(self, High):
        self.High = High
    
    def Low(self, Low):
        self.Low = Low
    
    def Close(self, Close):
        self.Close = Close
    
    def Volume(self, Volume):
        self.Volume = Volume
    
    def Did_Win(self, Win):
        self.Did_Win = Win
        
    def Profit(self, Profit):
        self.Profit = Profit
        
weekly_gold_entries_empty = []
for x in range(0, (data.shape[0])):
    temp = data.iloc[x]
    weekly_gold_entries_empty.append(temp)

weekly_gold_entries = [weekly_gold() for i in range(data.shape[0])]

for y in range(0, len(weekly_gold_entries)):
    weekly_gold_entries[y].Date = pd.Series(weekly_gold_entries_empty[y]).values[0]
    weekly_gold_entries[y].Open = pd.Series(weekly_gold_entries_empty[y]).values[1]
    weekly_gold_entries[y].High = pd.Series(weekly_gold_entries_empty[y]).values[2]
    weekly_gold_entries[y].Low = pd.Series(weekly_gold_entries_empty[y]).values[3]
    weekly_gold_entries[y].Close = pd.Series(weekly_gold_entries_empty[y]).values[4]
    weekly_gold_entries[y].Volume = pd.Series(weekly_gold_entries_empty[y]).values[5]
    weekly_gold_entries[y].Profit = pd.Series(weekly_gold_entries_empty[y]).values[6]
    weekly_gold_entries[y].Did_Win = pd.Series(weekly_gold_entries_empty[y]).values[7]
    
X_train, X_TstandVal = train_test_split(weekly_gold_entries, train_size = 0.6, test_size = 0.4, shuffle = False)
X_val, X_test = train_test_split(X_TstandVal, train_size =.5, test_size = .5, shuffle = False)


def print_Row(index):
    print("Date: " + weekly_gold_entries[index].Date)
    print("Open: " + str(weekly_gold_entries[index].Open))
    print("High: " + str(weekly_gold_entries[index].High))
    print("Low: " + str(weekly_gold_entries[index].Low))
    print("Close: " + str(weekly_gold_entries[index].Close))
    print("Volume: " + str(weekly_gold_entries[index].Volume))
    print("Profit: " + str(weekly_gold_entries[index].Profit))
    if(weekly_gold_entries[index].Did_Win == 0.0):
        print("Did_Win: Lose")
    else:
        print("Did_Win: Win")

#Discrete Category Calculations        
print("Discrete Category Calculations: \n")

class_P = 0
class_N = 0
winning_weeks_in_row = 0
losing_weeks_in_row = 0
three_weeks_winning = 0
three_weeks_losing = 0
for x in range(0, len(X_train)):
    if(X_train[x].Did_Win == 1):
        class_P += 1
        winning_weeks_in_row += 1
        losing_weeks_in_row = 0
    else:
        class_N += 1
        losing_weeks_in_row +=1
        winning_weeks_in_row = 0
    
    if(winning_weeks_in_row >= 3):
        three_weeks_winning += 1
    elif(losing_weeks_in_row >= 3):
        three_weeks_losing += 1
    else:
        continue
    
print("Number of times 3 winning weeks in a row occured: " + str(three_weeks_winning))
print("Number of times 3 losing weeks in a row occured: " + str(three_weeks_losing))

high_pos_change = 0
high_neg_change = 0
for y in range(0, len(X_train)):
    if(X_train[y].Profit >= (X_train[y].Close * 0.05)):
        high_pos_change += 1
    elif(X_train[y].Profit <= (-(X_train[y].Close) * 0.05)):
        high_neg_change += 1
    else:
        continue

print("Number of Weeks with greater than 5% win: " + str(high_pos_change))
print("Number of Weeks with greater than 5% loss: " + str(high_neg_change))

high_Volume = 0
closed_high = 0
for z in range(0, len(X_train)):
    if(X_train[z].Volume > (X_train[z].Close * 2)):
        high_Volume += 1
    
    if(X_train[z].Close > ((X_train[z].High + X_train[z].Low) / 2)):
        closed_high += 1
        
print("Number of Weeks with a high volume: " + str(high_Volume))
print("Number of Weeks that closed closer to the High than the Low: " + str(closed_high))
        

print("Number of Class Positives: " + str(class_P))
print("Number of Class Negatives: " + str(class_N))

print("\n\n")

#Contiguous Category Calculations
print("Contiguous Category Calculations: \n")

temp = []
for i in range(0, len(X_train)):
    temp.append((X_train[i].Profit / X_train[i].Close) * 100)
        
max_profit = max(temp)
min_profit = min(temp)
mean_profit = sum(temp) / len(temp)
median_profit = stats.median(temp)
std_profit = np.std(temp)
quartile_size = stats.quantiles(temp)


print("Max Profit from 1 week: " + str(max_profit))
print("Max Loss from 1 week: " + str(min_profit))
print("Total Mean Profit: " + str(mean_profit))
print("Total Median Profit: " + str(median_profit))
print("Profit Standard Deviation: " + str(std_profit))
print("Quartiles: Q1, Q2, and Q3 respectively: " + str(quartile_size))
    
#     if(weekly_profit > 0):
#         weekly_gold_entries[i].Did_Win = 1
#     else:
#         weekly_gold_entries[i].Did_Win = 0
    
#win_list = []
#win_amount = []
#for j in range(0, len(weekly_gold_entries)-1):
#    win_list.append(weekly_gold_entries[j].Did_Win)
#    win_amount.append(weekly_gold_entries[j].Profit)
#    
#csv_df = pd.DataFrame(win_list, win_amount)
#csv_df.to_csv('temp.csv')
```
